```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")
%matplotlib inline
```

```python
df = sns.load_dataset('penguins')

df.head()
```

```python
print("Shape:", df.shape)
df.info()
df.describe(include='all')
```

```python
df.isna().sum()
```

```python
df.to_csv('penguins_clean.csv', index=False)
```

```python
# 1. Bar Plot - Average body mass by species
plt.figure(figsize=(6,4))
sns.barplot(data=df, x="species", y="body_mass_g", estimator="mean", ci=None)
plt.title("Average Body Mass by Species")
plt.savefig("plots/barplot_species_mass.png")
```

```python
# 2. Box Plot - Flipper length by species
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="species", y="flipper_length_mm")
plt.title("Flipper Length Distribution by Species")
plt.savefig("plots/boxplot_species_flipper.png")
```

```python
# 3. Violin Plot - Body mass by island 
plt.figure(figsize=(6,4))
sns.violinplot(data=df, x="island", y="body_mass_g")
plt.title("Body Mass Distribution by Island")
plt.savefig("plots/violinplot_island_mass.png")
```

```python
# 4. KDE Plot - Distribution of flipper length
plt.figure(figsize=(6,4))
sns.kdeplot(data=df, x="flipper_length_mm", fill=True)
plt.title("KDE of Flipper Length")
plt.savefig("plots/kde_flipper.png")
```

```python
# 5. Swarm Plot - Body mass vs species
plt.figure(figsize=(6,4))
sns.swarmplot(data=df, x="species", y="body_mass_g")
plt.title("Swarm Plot of Body Mass by Species")
plt.savefig("plots/swarm_species_mass.png")
```

```python
# 6. Pair Plot - Pairwise relationships
sns.pairplot(df.dropna(), hue="species")
plt.savefig("plots/pairplot_species.png")
```

```python
# 7. Heatmap - Correlation between features
plt.figure(figsize=(6,4))
sns.heatmap(df.dropna().corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.savefig("plots/heatmap_corr.png")
```

```python
# 8. Histogram - Body mass
plt.figure(figsize=(6,4))
sns.histplot(df["body_mass_g"].dropna(), bins=20, kde=True)
plt.title("Histogram of Body Mass")
plt.savefig("plots/histogram_body_mass.png")
```

```python
# 9. Scatter Plot - Bill length vs flipper length
plt.figure(figsize=(6,4))
sns.scatterplot(data=df, x="bill_length_mm", y="flipper_length_mm", hue="species")
plt.title("Scatter Plot: Bill Length vs Flipper Length")
plt.savefig("plots/scatter_bill_flipper.png")
```

```python
# 10. Box Plot - Bill depth by island
plt.figure(figsize=(6,4))
sns.boxplot(data=df, x="island", y="bill_depth_mm")
plt.title("Bill Depth by Island")
plt.savefig("plots/boxplot_island_bill_depth.png")
```
